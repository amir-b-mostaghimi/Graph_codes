```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

import numpy as np
```

```python
# set the font name for a font family
plt.rcParams.update({'font.sans-serif':'Helvetica'})

fig= plt.figure(dpi=720)
ax= fig.add_axes([0.1,0.1,0.9,0.9])

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
    
plt.xlim([0, 8])
plt.ylim([2.9, 5.2])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=5)
ax.tick_params(which='minor', length=3)


# For the minor ticks, use no labels; default NullFormatter.
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

dx1=[0,8]
dy1=[4.22,4.22]
ax.plot(dx1,dy1, color='k',linestyle='dashed')

#####Ideal line######

dxi=[0,2,np.nan,3,5,np.nan,6,8] 
dyi=[4.22,4.22,np.nan,4.22,4.22,np.nan,4.22,4.22]
ax.plot(dxi,dyi,color= 'k',linewidth=4 )


#####The data######
cluster= [
    4.3743636,
3.92452989,
3.98061544
]
    
doped= [
4.72008309,
4.36507887,
3.6265375
]    

pure_metal= [
   3.02168964,
5.01649179 
]



for i in cluster:
    x=[3,5]
    y=[i,i]
    ax.plot(x,y,color='red',linewidth=4)
    x=[2,3,np.nan,5,6]
    y=[4.22,i,np.nan,i,4.22]
    ax.plot(x,y,color='red',linewidth=2,linestyle='dashed')
    
    
for i in doped:
    x=[3,5]
    y=[i,i]
    ax.plot(x,y,color='blue',linewidth=4)
    x=[2,3,np.nan,5,6]
    y=[4.22,i,np.nan,i,4.22]
    ax.plot(x,y,color='blue',linewidth=2,linestyle='dashed')
    
    
for i in pure_metal:
    x=[3,5]
    y=[i,i]
    ax.plot(x,y,color='green',linewidth=4)
    x=[2,3,np.nan,5,6]
    y=[4.22,i,np.nan,i,4.22]
    ax.plot(x,y,color='green',linewidth=2,linestyle='dashed')

    
ax.axes.get_xaxis().set_visible(False)
```
