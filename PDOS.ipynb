```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import pandas as pd

import numpy as np
```

```python
### set the font name for a font family
plt.rcParams.update({'font.sans-serif':'Helvetica'})

fig= plt.figure(dpi=720)
ax= fig.add_axes([0.1,0.1,0.9,0.9])

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
    

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylim((-25,250))
#plt.xlim((-7,5))

ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=5)
ax.tick_params(which='minor', length=3)


# For the minor ticks, use no labels; default NullFormatter.
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.axes.get_yaxis().set_visible(False)

##########

file_list=['Au.txt','CPd1.txt','DPd2.txt']

def dplot (center=1):
    for i in file_list:
        center= float(input('d-band center for %s' %(i)))
        data = pd.read_csv(i,sep="\s+", header = 0)
        data = pd.DataFrame(data)
        
        x = data[data.columns[0]]
        y = data[data.columns[1]]
        
        ##########################Lambda functions to adjust the label and the color of the lines based on the type of 
        #the input file
        label= lambda a: 'Au' if a=='Au.txt' else ( 'Pd₂' if a == 'DPd2.txt' else 'Pd₁')
        color= lambda a: 'green' if a=='Au.txt' else ( 'blue' if a == 'DPd2.txt' else 'red')
        ###########################
        
        t= ax.plot(x, y, label= label(i), color= color(i) ) # here color and label feed the name of the input files to the above functions to get the correct color and label
        color=t[0].get_color()
        ax.plot([center,center],[-25,2255],linestyle='dashed', color=color)
        
        ax.legend()
        
ax.plot([0,0],[-25,2500],linestyle='dashed',color='k')
dplot()
```
