```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
# set the font name for a font family
plt.rcParams.update({'font.sans-serif':'Helvetica'})

fig= plt.figure(dpi=720)
ax= fig.add_axes([0.1,0.1,0.9,0.9])

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
    
plt.xlim([3.52, 4.93])
plt.ylim([0, 0.8])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=5)
ax.tick_params(which='minor', length=3)


# For the minor ticks, use no labels; default NullFormatter.
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))

dx1= [3.52,4.22,4.93]
dy1= [0,0.7,0]

dx2=[0,5]
dy2=[0.7,0.7]

ax.plot(dx1,dy1, color='k')
ax.plot(dx2,dy2, color='k',linestyle='dashed')




ax.plot(4.72008309, 0.19991691, marker="s", markersize=15, color="blue", label ='Doped',linestyle = 'None')
ax.plot(4.36507887, 0.55492113, marker="s", markersize=15, color="blue")
ax.plot(3.6265375, 0.1065375, marker="s", markersize=15, color="blue")

ax.plot(4.37, 0.55, marker="s", markersize=15, color="red", label ='Cluster',linestyle = 'None')
ax.plot(3.92, 0.40, marker="s", markersize=15, color="red")
ax.plot(3.98,0.46 , marker="s", markersize=15, color="red")


leg = ax.legend(loc ='upper right',prop={'size': 14})
```
